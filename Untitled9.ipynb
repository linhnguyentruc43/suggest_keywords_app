```python
# H·ªÜ TH·ªêNG G·ª¢I √ù T·ª™ KH√ìA T√åM KI·∫æM S·ª¨ D·ª§NG TF-IDF
# ƒê√¢y l√† notebook h∆∞·ªõng d·∫´n chi ti·∫øt t·ª´ng b∆∞·ªõc x√¢y d·ª±ng h·ªá th·ªëng g·ª£i √Ω t·ª´ kh√≥a s·ª≠ d·ª•ng TF-IDF v√† Cosine Similarity b·∫±ng Python
# üîπ B∆Ø·ªöC 1: C√ÄI ƒê·∫∂T V√Ä IMPORT TH∆Ø VI·ªÜN
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# üîπ B∆Ø·ªöC 2: T·∫¢I D·ªÆ LI·ªÜU
from google.colab import files
uploaded = files.upload()
# üîπ B∆Ø·ªöC 3: ƒê·ªåC D·ªÆ LI·ªÜU
df = pd.read_csv(list(uploaded.keys())[0])
df = df.dropna(subset=['keyword'])
df['keyword'] = df['keyword'].astype(str).str.lower()
df['keyword'] = df['keyword'].str.replace(r'[^\w\s]', '', regex=True)
df.head()
# üîπ B∆Ø·ªöC 4: BI·∫æN T·ª™ KH√ìA TH√ÄNH VECTOR TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['keyword'])
# üîπ B∆Ø·ªöC 5: VI·∫æT H√ÄM G·ª¢I √ù
def suggest_keywords(input_text, top_n=5):
    input_vec = vectorizer.transform([input_text])
    cosine_sim = cosine_similarity(input_vec, X).flatten()
    indices = cosine_sim.argsort()[-top_n:][::-1]
    return df['keyword'].iloc[indices]
# üîπ B∆Ø·ªöC 6: TH·ª¨ NGHI·ªÜM
input_text = "truyen ngon tinh"
suggest_keywords(input_text)
```
